```python
import pandas as pd
import os
import altair as alt
from datetime import datetime

# First, define the variable run_date to be the current time
run_date = datetime.now()

# Second, define a list 'all_symbol_data', which will be empty
all_symbol_data = []

# Iterate over the files in the specified directory
for file_name in os.listdir('/home/stefan/Desktop/STADIUM-DATA/END_RESULT_EFFICIENCY'):
    file_path = os.path.join('/home/stefan/Desktop/STADIUM-DATA/END_RESULT_EFFICIENCY', file_name)
    if os.path.isfile(file_path):
        # For each file, make a dataframe, ldf
        ldf = pd.read_csv(file_path)
        
        # For each ldf, get the mean of the 'hours' column, as variable mean_duration
        mean_duration = ldf['hours'].mean()
        
        # The unique value of the 'symbol' column, assign to variable 'symbol'
        symbol = ldf['symbol'].unique()[0]
        
        # Create a dictionary object, dict, with the specified key-value pairs
        data_dict = {
            'mean_duration': mean_duration,
            'symbol': symbol,
            'run_date': run_date
        }
        
        # Append the dictionary object to the list, all_symbol_data
        all_symbol_data.append(data_dict)

# Once the loop terminates, create a dataframe 'asd', with data as the list all_symbol_data
asd = pd.DataFrame(all_symbol_data)

# Create a histogram visualization from Altair
chart = alt.Chart(asd).mark_bar().encode(
    x='mean_duration',
    y='count()',
    color=alt.Color('mean_duration', bin=True)
)

# Display the chart
chart.show()
```

```python
# Create an interactive histogram visualization from Altair
chart = alt.Chart(asd).mark_bar().encode(
    x='mean_duration',
    y='count()',
    color=alt.Color('mean_duration', bin=True),
    # Add tooltip to display 'symbol' when hovering over a bar
    tooltip=['symbol', 'mean_duration']
).interactive()  # Make the chart interactive

# Display the chart
chart.show()
```
