```python
import altair as alt
import os
import pandas as pd
```

```python
# folder_loc = '/home/stefan/Desktop/STADIUM_DATA/ALL_END_PREDICTIONS_TIME_DURATIONS'


# Define the directory to iterate
directory = '/home/stefan/Desktop/STADIUM_DATA/ALL_END_PREDICTIONS_TIME_DURATIONS'
# Initialize an empty DataFrame to store the combined data
combined_df = pd.DataFrame()

# List of small DataFrames (for illustration purposes)
small_dfs = []

# Iterate over the files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        # Add the file name to the list
        # files.append(filename)
        
        # Create a dataframe for each file
        df = pd.read_csv(os.path.join(directory, filename))
        small_dfs.append(df)
       
# Loop through each small DataFrame and add it as a new row to the combined DataFrame
for df in small_dfs:
    combined_df = pd.concat([combined_df, df], ignore_index=True)

# Print the resulting combined DataFrame
print(combined_df)

```

```python
# Create the Altair chart
chart = alt.Chart(combined_df).mark_bar().encode( 
    x=alt.X('commence:T', axis=alt.Axis(labelAngle=-45)),
    y='hours:Q',
    color='hours:Q',
    tooltip=['symbol','commence:T']
).properties(
    width=900,
    title={
        "text": "HFT MLOPS Prediction Results, by Time",
        "subtitle": "The time, in hours required for each prediction to go from buy to sell, as a time series"
    }
).interactive()

# Display the chart
chart
```

load all duration-fitted predictions as a dashboard, from [SCP-PRED directory, sent to trader facility](SCP-PRED)

```python
directory = '/home/stefan/Desktop/STADIUM_DATA/SCP-PRED/'
# Initialize an empty DataFrame to store the combined data
recent_predictions_duration_fitted = pd.DataFrame()
# List of small DataFrames (for illustration purposes)
small_dfs_2 = []
# Iterate over the files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        extracted_part = filename.split('-')[0] # Extracting the part before the first '-'
        df = pd.read_csv(os.path.join(directory, filename), header=None)
        df.columns = ['time', 'y_pred', 'method', 'a', 'b', 'c', 'd', 'e', 'length', 'bin']
        df['symbol'] = extracted_part
        small_dfs_2.append(df)
for df in small_dfs_2:
    recent_predictions_duration_fitted = pd.concat([recent_predictions_duration_fitted, df], ignore_index=True)
print(recent_predictions_duration_fitted)
recent_predictions_duration_fitted = recent_predictions_duration_fitted.sort_values(by='time', ascending=True)
```

```python
recent_predictions_duration_fitted.columns
```

```python
# Create the Altair chart
chart = alt.Chart(recent_predictions_duration_fitted.tail(50)).mark_point().encode(
    x=alt.X('time',axis=alt.Axis(labelAngle=-45)),
    y='symbol',
    tooltip=['symbol','time']
).properties(
    title='Most Recent Trade Predictions'
)

# Display the chart
chart.show()
```
